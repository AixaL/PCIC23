```python
ruta_archivo_json = "/home/ar/Data/AixaPL/Version1/PCIC23/downloadFiles/CasualConversationsV2.json"
datos = pd.read_json(ruta_archivo_json, orient='records')
```

```python
datos.head()
```

```python
current_directory = '/home/ar/Data/AixaPL/Version1/PCIC23/downloadFiles/'
dataset = []
# current_directory = os.getcwd()
with os.scandir(current_directory) as ficheros:
    subdirectorios = [fichero.name for fichero in ficheros if fichero.is_dir()]

for directorio in subdirectorios:
    # if 'CCv2' not in directorio and 'error' not in directorio:
    if 'CCv2' in directorio:
        print(os.path)
        print(directorio)
        audio_files = [str(file) for file in Path(os.path.join(current_directory, directorio)).iterdir() if file.name.endswith(".wav")]
        # audio_files = [os.path.join(directorio, file) for file in os.listdir(directorio) if file.endswith(".wav")]
        print(audio_files)
        for file in audio_files:
            fileName= os.path.basename(file).split("-")[0]
            fileName = fileName.replace("wav","mp4")

            dataJson = datos[datos['video_name'] == fileName]

            if not dataJson.empty:
                entry = {
                    'file_path': file,
                    'file_name': fileName,
                    'subject_id': dataJson['subject_id'].values[0],
                    'age': dataJson['age'].values[0],
                    'gender': dataJson['gender'].values[0],
                    'native_language': dataJson['native_language'].values[0]
                }
                dataset.append(entry)

dataset_df = pd.DataFrame(dataset)
```
